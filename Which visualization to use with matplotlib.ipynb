<h2 style="color:blue;"> Line plot :</h2> 

**Function :**

displays information as a series of data points connected by straight line segments.

**When to use it ?**

Line plots are generally used to visualize the directional movement of one or more data over time. In this case, the X axis would be datetime and the y axis contains the measured quantity, like, stock price, weather, monthly sales, etc.

A line plot is often the first plot of choice to visualize any time series data.

````python
# for this plot I will just use an example out of the provided data since there is no dates
data = pd.DataFrame({
   'Price': [4, 7, 20, 18, 30],
    'Sales':[10, 12, 8, 14, 20],
   }, index=[1990, 1997, 2003, 2009, 2014])
plt.plot(data.Price)
````

<h2 style="color:blue;"> Area plot :</h2>

**Function :**

Commonly used to represent cumulated totals using numbers or percentages over time.

**When to use it?**

It is used when we try to compare two or more quantities over the time.

````python
plt.fill_between(data.index, data.Sales, label = 'Sales')
plt.fill_between(data.index, data.Price, alpha = 0.4, label = 'Price')
plt.legend()
````

<h2 style="color:blue;">Histogram</h2>

**Function :**

It plots a general distribution of a continuous variable’s values as a series of bars by dividing the data into a number of bins. It then plots the frequency data points in each bin (i.e. the interval of points). 

**When to use:**

 When we need the count of data ranges of a variable in a plot.
 
 

````python
plt.figure(figsize=(12,6))
plt.hist(df.Age,
        facecolor = 'orange',
        edgecolor = 'DodgerBlue',
        bins = 20)
````

<h2 style="color:blue;">Bar Chart :</h2>

**When to use:**

It is used when to compare between several groups.

Bar Plot shows the distribution of data over several groups. It is commonly confused with a histogram which only takes numerical data for plotting. It helps in comparing multiple numeric values.


````python
embaerkedNB = df.groupby('Embarked').count()['PassengerId']

plt.figure(figsize=(8,6))
x = range(3)
plt.bar(x, embaerkedNB, color='green')
plt.xticks(x, embaerkedNB.index)
````

<h2 style="color:blue;">Stacked Bar Chart :</h2>

**Function :**

The stacked bar chart (aka stacked bar graph) extends the standard bar chart from looking at numeric values across one categorical variable to two. Each bar in a standard bar chart is divided into a number of sub-bars stacked end to end, each one corresponding to a level of the second categorical variable.

**When to use it ?**

When we are looking for comparing numeric values between levels of a categorical variable.

We want to move to a stacked bar chart when we care about the relative decomposition of each primary bar based on the levels of a second categorical variable. 

````python
pclassNB = df.groupby(df.Pclass).count()['PassengerId']
pclassNB
````

````python
plt.bar(x, embaerkedNB, color='orange')
plt.bar(x, pclassNB, color='yellow', bottom = embaerkedNB)

plt.show()
````

<h2 style="color:blue;">Pie Chart :</h2>

**Fuction :**

It illustrates nmerical proportions.

**When to use:** 

Pie chart should be used seldom used as It is difficult to compare sections of the chart. Bar plot is used instead as comparing sections is easy.

display the proportion of each value against the total sum of values. 


````python
plt.figure(figsize=(10,5))
plt.pie(embaerkedNB,
       labels = embaerkedNB.index, autopct = '%1.1f%%',
       explode = [0,0,0.1])

plt.axis('equal')
````

<h2 style="color:blue;">Scatter plot :</h2>

**When to use:**

It is used in Machine learning concepts like regression, where x and y are continuous variables. It is also used in clustering scatters or outlier detection.

 It realy helps in detecting outliers in the plot.
 

````python
plt.scatter(df.Fare, df.Age, color = 'pink')
````

<h2 style="color:blue;">Bubble plot :</h2> 

**Function :**

Variation of the scatter plot, that displays three dimensions of data (x,y,z), the size of bubbles is determined by z. 
It displays data as a cluster of circles.

**When to use it?**

When you are looking for making the comparison between three or four (the color) variables rather than just two. 

````python
plt.figure(figsize=(10,6))
plt.scatter('Fare', 'Age',
             s=df.Pclass * 100,
             alpha=0.4, 
             c = 'Survived',
             data=df)
````

<h2 style="color:blue;">Box plot :</h2>

**When to use:**

It should be used when we require to use the overall statistical information on the distribution of the data. It can be used to detect outliers in the data.

helps in understanding our distribution better.

A caveat of using boxplot is the number of observations in the unique value is not defined, Jitter Plot in Seaborn can overcome this caveat or Violinplot is also useful

````python
plt.figure(figsize=(10,6))
plt.boxplot(df.Fare)
````

<h2 style="color:blue;">Violin plot :</h2> 

**Function :**

It is an extension to boxplot which shows the probability density of the data at different values.

**When to use:**

It should be used when we require a better intuitive understanding of data, because it gives a much broader understanding of the distribution. 

It resembles a violin and dense areas point the more distribution of data otherwise hidden by box plots(see the image).

It also serves a lot for comparison of distributions between multiple groups. 



<img src="https://blog.bioturing.com/wp-content/uploads/2018/11/BoxViolin.gif" width="700"></img>

````python
plt.violinplot(df.Fare)
````

<h2 style="color:blue;">Error Bar Chart :</h2> 

**Function :**

Help you indicate estimated error or uncertainty to give a general sense of how precise a measurement is this is done through the use of markers drawn over the original graph and its data points. 

**When to use it?**

In experimental design, the measurements lack perfect precision. So, we have to repeat the measurements. It results in obtaining a set of values. The representation of the distribution of data values is done by plotting a single data point (known as mean value of dataset) and an error bar to represent the overall distribution of data.

````python
plt.figure(figsize=(10,5))
plt.errorbar(df.PassengerId[:40], df.Fare[:40],  yerr=df.Age[:40], fmt = '.-')
````

<h2 style="color:blue;">Waffle Chart :</h2>

**Function :**

created to display and visualize data in relationship to a whole.

**When to use it?**

When you are looking for displaying progress towards goals. 

````python
#!pip install pywaffle
from pywaffle import Waffle

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

dff = pd.DataFrame(pclassNB)

fig = plt.figure(
   FigureClass=Waffle,
   rows=15,
   values=dff.PassengerId,
   labels=list(dff.index)
)
plt.show()
````
